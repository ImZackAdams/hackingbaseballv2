```python
from geopy.geocoders import Nominatim
import pandas as pd

# Convert the stadiums dictionary to a DataFrame for easier processing
df_stadiums = pd.DataFrame(list(stadiums.items()), columns=['Team', 'Stadium_Address'])

# Initialize the geocoder
geolocator = Nominatim(user_agent="mlb_geocoder")

# Function to geocode addresses
def geocode_address(address):
    try:
        location = geolocator.geocode(address)
        return pd.Series([location.latitude, location.longitude])
    except Exception as e:
        print(f"Error geocoding {address}: {e}")
        return pd.Series([None, None])

# Apply the function to geocode each stadium address
df_stadiums[['Latitude', 'Longitude']] = df_stadiums['Stadium_Address'].apply(geocode_address)
```
